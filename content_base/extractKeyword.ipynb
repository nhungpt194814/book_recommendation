```python
import pandas as pd
import numpy as np
import pandas as pd
import numpy as np
from nltk.corpus import stopwords
import nltk

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
```

```python
# Reading the file
data = pd.read_csv("clean.csv")
```

```python
# keep the data that we need
data = data.drop(columns=['numPages','avgRating','ratingHistogram','publishYear','ratingsCount','reviewsCount'])
```

```python
# array to string
data['author'] = data['author'].astype(str)
data['characters'] = data['characters'].astype(str)
data['genres'] = data['genres'].astype(str)
data['series'] = data['series'].astype(str)
data['description'] = data['description'].astype(str)
```

```python
# Remove punctuations
import string
punctuations = string.punctuation
def remove_punctuations(text):
    return text.translate(str.maketrans('', '', punctuations))

data.genres = data.genres.apply(remove_punctuations)
data.series = data.series.apply(remove_punctuations)
data.characters = data.characters.apply(remove_punctuations)
data.author = data.author.apply(remove_punctuations)
data.description=data.description.apply(func = remove_punctuations)
```

```python
# convert the lettercase to lower
data[["title", "author", "publisher", "description","series","genres","characters"]] = pd.concat([data[col].astype(str).str.lower().str.strip() for col in ["title", "author", "publisher", "description","series","genres","characters"]], axis=1)
data.head()
```

```python
# Transform author names, publisher into single token
data["author2"] = data["author"].str.strip().str.replace(' ','_')
data["publisher2"] = data["publisher"].str.strip().str.replace(' ','_')
```

```python
import nltk
nltk.download('stopwords')
# Function for removing stop words
def remove_stop_words(text):
    text = text.split()
    stops = set(stopwords.words("english"))
    text = [w for w in text if not w in stops]
    text = " ".join(text)
    return text
data['description']=data.description.apply(func = remove_stop_words)
```

```python
# lemmatize (ran -> run)
def lemma(text):
    lmtzr = WordNetLemmatizer()
    text = text.split()
    text = [lmtzr.lemmatize(word) for word in text]
    return ' '.join(text)

data['lemmatized_description'] = data['description'].apply(func = lemma)
data.head()
```

```python
# extract keyword with keyBERT
!pip install keybert
from keybert import KeyBERT
kw_model = KeyBERT()

def get_keywords(text):
    keywords = kw_model.extract_keywords(text, keyphrase_ngram_range=(1, 1), stop_words="english")
    keywords = " ".join([k[0] for k in keywords])
    return keywords
```

```python
data["keywords"] = data.lemmatized_description.apply(get_keywords)
data.keywords.head()
```

```python
# Merge all textual information to create summary
data["sum"] = data[["title", 'author2', 'publisher2', 'keywords','language','series','characters','genres']].fillna('').agg(' '.join, axis=1)
data = data.drop(columns=['author2', 'publisher2'])

data.head()
```

```python
data = data.drop(columns=['lemmatized_description','keywords','description','imageUrl','genres','publisher','author','characters','language','series'])
data.head(3)
```

```python
data.to_csv("keywords.csv", sep=",", index=False)
```
